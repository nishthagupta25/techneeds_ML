# **Dataset1=Car_Price**

```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

# **Multiple linear regression :**

```python
#printing dataset1
carprice=pd.read_csv("CarPrice_Assignment.csv")
carprice
```

```python
#printing only two columns
carprice=carprice[["enginesize","price"]]
carprice
```

```python
x=carprice["enginesize"]
y=carprice["price"]
```

```python
#plotting the graph :
plt.scatter(x,y)
plt.xlabel("Car Engine Size")
plt.ylabel("Car Price")
plt.show
```

# **For training and testing the model :**

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.3 , random_state=23)
```

```python
x_train #It takes random rows for traing data
```

```python
x_train=np.array(x_train).reshape(-1,1)
x_train
```

```python
x_test=np.array(x_test).reshape(-1,1)
x_test
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
c= lr.intercept_
c
```

```python
m= lr.coef_
m
```

```python
y_pred_train=m*x_train + c
y_pred_train.flatten()
```

```python
y_pred_train1= lr.predict(x_train)
y_pred_train1
```

```python
plt.scatter(x_train,y_train)
plt.plot(x_train,y_pred_train1,color="red")
plt.xlabel(" petal length")
plt.ylabel(" petal width")
```

```python
y_pred_test1= lr.predict(x_test)
y_pred_test1
```

```python
plt.scatter(x_test,y_test)
plt.plot(x_test,y_pred_test1,color="red")
plt.xlabel(" petal length")
plt.ylabel(" petal width")
```

# **Evaluate the Model :**

```python
from sklearn.metrics import r2_score
r2_score=(y_test,y_pred_test1)
```

# **Plot the Results:**

```python
plt.figure(figsize=(10,5))
plt.scatter(y_test,y_pred_test1)
plt.xlabel("Actual")
plt.ylabel("predicted")
plt.title("Actual vs predicted")

```

# **Predicted Values**

```python
pred_y_df = pd.DataFrame({'Actual Value':y_test,'Predicted Value':y_pred_test1,'Difference':y_test - y_pred_test1})
pred_y_df.head(20)
```
